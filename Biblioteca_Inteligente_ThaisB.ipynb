```python
import os
import json
from fpdf import FPDF

# -------------------------------------
# SISTEMA DE BIBLIOTECA DIGITAL - THAIS B.
# -------------------------------------

# Caminho da pasta da biblioteca
biblioteca_path = "/content/biblioteca"

# Criar pasta se n√£o existir
if not os.path.exists(biblioteca_path):
    os.mkdir(biblioteca_path)

# Criar cat√°logo inicial (se ainda n√£o existir)
catalogo = {
    "catalogo": [
        {"id": 1, "titulo": "Dom Casmurro", "autor": "Machado de Assis", "ano": 1899, "arquivo": "Dom_Casmurro.pdf"},
        {"id": 2, "titulo": "Pregui√ßa e Filosofia", "autor": "John Perry", "ano": 2011, "arquivo": "Preguica_e_Filosofia.pdf"},
        {"id": 3, "titulo": "A Origem das Esp√©cies", "autor": "Charles Darwin", "ano": 1859, "arquivo": "A_Origem_das_Especies.pdf"}
    ]
}

# Criar arquivo JSON
if not os.path.exists(f"{biblioteca_path}/catalogo.json"):
    with open(f"{biblioteca_path}/catalogo.json", "w", encoding="utf-8") as f:
        json.dump(catalogo, f, indent=4, ensure_ascii=False)

# Criar PDFs simples de exemplo
for livro in catalogo["catalogo"]:
    caminho_pdf = f"{biblioteca_path}/{livro['arquivo']}"
    if not os.path.exists(caminho_pdf):
        pdf = FPDF()
        pdf.add_page()
        pdf.set_font("Arial", size=12)
        pdf.cell(0, 10, f"T√≠tulo: {livro['titulo']}", ln=True)
        pdf.cell(0, 10, f"Autor: {livro['autor']}", ln=True)
        pdf.cell(0, 10, f"Ano: {livro['ano']}", ln=True)
        pdf.output(caminho_pdf)

print("‚úÖ Biblioteca inicializada com sucesso!")

# -------------------------------------
# FUN√á√ïES
# -------------------------------------

def carregar_catalogo():
    with open(f"{biblioteca_path}/catalogo.json", "r", encoding="utf-8") as f:
        return json.load(f)

def salvar_catalogo(dados):
    with open(f"{biblioteca_path}/catalogo.json", "w", encoding="utf-8") as f:
        json.dump(dados, f, indent=4, ensure_ascii=False)

def listar_documentos():
    dados = carregar_catalogo()
    print("\nüìö Documentos dispon√≠veis:")
    for livro in dados["catalogo"]:
        print(f"ID: {livro['id']} | {livro['titulo']} - {livro['autor']} ({livro['ano']})")

    print("\nüìÇ Arquivos por tipo:")
    for arquivo in os.listdir(biblioteca_path):
        if arquivo.endswith(".pdf"):
            print("  - PDF:", arquivo)
        elif arquivo.endswith(".json"):
            print("  - JSON:", arquivo)

    print("\nüìÖ Organiza√ß√£o por ano:")
    ordenados = sorted(dados["catalogo"], key=lambda x: x["ano"])
    for livro in ordenados:
        print(f"  {livro['ano']} - {livro['titulo']}")

def adicionar_documento():
    dados = carregar_catalogo()
    novo_id = max([livro["id"] for livro in dados["catalogo"]]) + 1
    titulo = input("T√≠tulo do novo documento: ")
    autor = input("Autor: ")
    ano = int(input("Ano de publica√ß√£o: "))
    arquivo = titulo.replace(" ", "_") + ".pdf"

    # Criar PDF b√°sico
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)
    pdf.cell(0, 10, f"T√≠tulo: {titulo}", ln=True)
    pdf.cell(0, 10, f"Autor: {autor}", ln=True)
    pdf.cell(0, 10, f"Ano: {ano}", ln=True)
    pdf.output(f"{biblioteca_path}/{arquivo}")

    dados["catalogo"].append({"id": novo_id, "titulo": titulo, "autor": autor, "ano": ano, "arquivo": arquivo})
    salvar_catalogo(dados)
    print("‚úÖ Documento adicionado com sucesso!")

def renomear_documento():
    nome_antigo = input("Nome do arquivo atual (ex: Dom_Casmurro.pdf): ")
    nome_novo = input("Novo nome (ex: DomCasmurroNovo.pdf): ")
    try:
        os.rename(f"{biblioteca_path}/{nome_antigo}", f"{biblioteca_path}/{nome_novo}")
        dados = carregar_catalogo()
        for livro in dados["catalogo"]:
            if livro["arquivo"] == nome_antigo:
                livro["arquivo"] = nome_novo
        salvar_catalogo(dados)
        print("‚úÖ Documento renomeado!")
    except FileNotFoundError:
        print("‚ö†Ô∏è Arquivo n√£o encontrado.")

def remover_documento():
    nome = input("Nome do arquivo que deseja remover (ex: Dom_Casmurro.pdf): ")
    caminho = f"{biblioteca_path}/{nome}"
    if os.path.exists(caminho):
        os.remove(caminho)
        dados = carregar_catalogo()
        dados["catalogo"] = [livro for livro in dados["catalogo"] if livro["arquivo"] != nome]
        salvar_catalogo(dados)
        print("üóëÔ∏è Documento removido!")
    else:
        print("‚ö†Ô∏è Arquivo n√£o encontrado.")

# -------------------------------------
# MENU PRINCIPAL
# -------------------------------------

while True:
    print("\n=== MENU BIBLIOTECA ===")
    print("1 - Listar documentos")
    print("2 - Adicionar documento")
    print("3 - Renomear documento")
    print("4 - Remover documento")
    print("5 - Sair")

    opcao = input("Escolha uma op√ß√£o: ")

    if opcao == "1":
        listar_documentos()
    elif opcao == "2":
        adicionar_documento()
    elif opcao == "3":
        renomear_documento()
    elif opcao == "4":
        remover_documento()
    elif opcao == "5":
        print("üëã Saindo da biblioteca...")
        break
    else:
        print("‚ö†Ô∏è Op√ß√£o inv√°lida. Tente novamente.")
```
