```python
import os
import json
from fpdf import FPDF

# Criei a pasta da biblioteca
pasta_biblioteca = "/content/biblioteca"

if not os.path.exists(pasta_biblioteca):
    os.mkdir(pasta_biblioteca)

# Criei um catálogo fictício inicial
catalogo = {
    "catalogo": [
        {"id": 1, "titulo": "Dom Casmurro", "autor": "Machado de Assis", "ano": 1899, "arquivo": "Dom_Casmurro.pdf"},
        {"id": 2, "titulo": "Preguiça e Filosofia", "autor": "John Perry", "ano": 2011, "arquivo": "Preguica_e_Filosofia.pdf"},
        {"id": 3, "titulo": "A Origem das Espécies", "autor": "Charles Darwin", "ano": 1859, "arquivo": "A_Origem_das_Especies.pdf"}
    ]
}

# Salvei o catálogo em JSON
caminho_catalogo = f"{pasta_biblioteca}/catalogo.json"
if not os.path.exists(caminho_catalogo):
    with open(caminho_catalogo, "w", encoding="utf-8") as f:
        json.dump(catalogo, f, indent=4, ensure_ascii=False)

# Peguei oslivros em PDF simples e criei a biblioteca
for livro in catalogo["catalogo"]:
    caminho_pdf = f"{pasta_biblioteca}/{livro['arquivo']}"
    if not os.path.exists(caminho_pdf):
        pdf = FPDF()
        pdf.add_page()
        pdf.set_font("Arial", size=12)
        pdf.cell(0, 10, f"Título: {livro['titulo']}", ln=True)
        pdf.cell(0, 10, f"Autor: {livro['autor']}", ln=True)
        pdf.cell(0, 10, f"Ano: {livro['ano']}", ln=True)
        pdf.output(caminho_pdf)

print("Biblioteca criada com sucesso.")
```

```python
import os
import json

# Criei o caminho dos arquivos
pasta_biblioteca = "/content/biblioteca"
caminho_catalogo = f"{pasta_biblioteca}/catalogo.json"

# criei uma função para carregar o catálogo
def carregar_catalogo():
    with open(caminho_catalogo, "r", encoding="utf-8") as f:
        return json.load(f)

# agora outra função para salvar o catálogo
def salvar_catalogo(dados):
    with open(caminho_catalogo, "w", encoding="utf-8") as f:
        json.dump(dados, f, indent=4, ensure_ascii=False)

# Iniciando confecção por partes da interface

# Listar documentos
def listar_documentos():
    dados = carregar_catalogo()
    print("\nDocumentos disponíveis:")
    for livro in dados["catalogo"]:
        print(f"{livro['id']} - {livro['titulo']} ({livro['autor']}, {livro['ano']})")

# Adicionar novo documento
def adicionar_documento():
    dados = carregar_catalogo()
    novo_id = max([livro["id"] for livro in dados["catalogo"]]) + 1
    titulo = input("Título: ")
    autor = input("Autor: ")
    ano = int(input("Ano de publicação: "))
    arquivo = titulo.replace(" ", "_") + ".pdf"

    novo_livro = {"id": novo_id, "titulo": titulo, "autor": autor, "ano": ano, "arquivo": arquivo}
    dados["catalogo"].append(novo_livro)
    salvar_catalogo(dados)

    print("Novo documento adicionado com sucesso.")

# Renomear arquivo existente
def renomear_documento():
    nome_antigo = input("Nome atual do arquivo (ex: Dom_Casmurro.pdf): ")
    nome_novo = input("Novo nome (ex: DomCasmurroNovo.pdf): ")

    caminho_antigo = f"{pasta_biblioteca}/{nome_antigo}"
    caminho_novo = f"{pasta_biblioteca}/{nome_novo}"

    if os.path.exists(caminho_antigo):
        os.rename(caminho_antigo, caminho_novo)
        dados = carregar_catalogo()
        for livro in dados["catalogo"]:
            if livro["arquivo"] == nome_antigo:
                livro["arquivo"] = nome_novo
        salvar_catalogo(dados)
        print("Arquivo renomeado com sucesso.")
    else:
        print("Arquivo não encontrado.")

# Remover arquivo
def remover_documento():
    nome = input("Nome do arquivo que deseja remover (ex: Dom_Casmurro.pdf): ")
    caminho = f"{pasta_biblioteca}/{nome}"

    if os.path.exists(caminho):
        os.remove(caminho)
        dados = carregar_catalogo()
        dados["catalogo"] = [livro for livro in dados["catalogo"] if livro["arquivo"] != nome]
        salvar_catalogo(dados)
        print("Documento removido.")
    else:
        print("Arquivo não encontrado.")

# Criando o menu principal
while True:
    print("\n==== MENU BIBLIOTECA ====")
    print("1 - Listar documentos")
    print("2 - Adicionar documento")
    print("3 - Renomear documento")
    print("4 - Remover documento")
    print("5 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        listar_documentos()
    elif opcao == "2":
        adicionar_documento()
    elif opcao == "3":
        renomear_documento()
    elif opcao == "4":
        remover_documento()
    elif opcao == "5":
        print("Saindo do sistema.")
        break
    else:
        print("Opção inválida.")
```
